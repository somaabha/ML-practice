## Gradient Descent

```python
import numpy as np
import pandas as pd
import math
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    m=b=0
    learning_rate=0.0002
    iterations=1000
    n=len(x)
    costprev=0
    
    for i in range(iterations):
        y_pred=m*x+b
        md=-(2/n)*sum(x*(y-y_pred))
        bd=-(2/n)*sum(y-y_pred)
        cost=(1/n)*sum((y-y_pred)**2)
        m=m-learning_rate*md
        b=b-learning_rate*bd
        if math.isclose(cost, costprev, rel_tol=1e-20):
            break
        costprev=cost
        print("m {}, b{}, i {}, cost {}".format(m,b,iterations,cost))
    return m,b
```

```python
def predict_using_sklearn():
    df = pd.read_csv("C:/Users/somaa/Downloads/test_scores.csv")
    r = LinearRegression()
    r.fit(df[['math']],df.cs)
    return r.coef_, r.intercept_
```

```python
df = pd.read_csv("C:/Users/somaa/Downloads/test_scores.csv")
x = np.array(df.math)
y = np.array(df.cs)
```

```python
m, b = gradient_descent(x,y)
print("Using gradient descent function: Coef {} Intercept {}".format(m, b))
```

```python
m_sklearn, b_sklearn = predict_using_sklearn()
print("Using sklearn: Coef {} Intercept {}".format(m_sklearn,b_sklearn))
```

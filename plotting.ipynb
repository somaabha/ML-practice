```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defing the functions
def sin(x):
    y=np.sin(x)
    return y

def cos(x):
    y=np.cos(x)
    return y
```

```python
c=np.linspace(0,360,100)
c=np.deg2rad(c)

y=sin(c)
y1=cos(c)
```

```python
plt.plot(c,y,label="y=sin(x)")
plt.title("Sin x")
plt.xlim(0,6)
plt.ylim(-1,1)
plt.legend()
plt.grid()
```

```python
#subplots
plt.subplot(1,2,1)
plt.plot(c, y, color="purple", label="y=sin(x)")
plt.xlabel("x [degree]", size=10)
plt.ylabel("y", size=10)
plt.title("Sine Function", size=18, pad=20)
plt.xlim(0,6)
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(c, y1, color="green", label="cos(x)")
plt.title("Cosine Function", size=18, pad=20)
plt.xlim(0,6)
plt.legend()

plt.tight_layout()
plt.grid()
plt.show()
```

```python
#generating noise
noise=np.random.random(100)
noise
```

```python
#plotting noise
plt.plot(c, y, color="purple", label="y=sin(x)")
plt.plot(c, y+noise, color="green", label="noise")
plt.xlim(0,6)
plt.grid()
plt.legend()
```

```python
#subplots with noise
plt.subplot(1,2,1)
plt.plot(c, y+noise, color="purple", label="y=sin(x)")
plt.xlabel("x [degree]", size=10)
plt.ylabel("y", size=10)
plt.title("Sine Function", size=18, pad=20)
plt.xlim(0,6)
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(c, y1+noise, color="green", label="cos(x)")
plt.title("Cosine Function", size=18, pad=20)
plt.xlim(0,6)
plt.legend()

plt.tight_layout()
plt.grid()
plt.show()
```

```python
#curve fitting using scipy
x = np.linspace(0, 50, 100)
def polynomial(x):
    return 0.5 * (x**2) + 0.7 * x + 10
y = polynomial(x)
noise = np.random.random(100) * 100
ydata = y + noise

plt.scatter(x, ydata, color="red")
plt.show()
```

```python
from scipy.optimize import curve_fit

def polynomial(x, a, b, c):
  y = a * x**2 + b * x + c
  return y

# x and y data are our data, that hasn't been fitted
curve_fit(polynomial, x, ydata)
```

```python
x = np.linspace(0, 50, 100)
noise = np.random.random(100) * 100
ydata = y + noise
yfit = polynomial(x, a=0.49507611, b=1.16682362, c=52.62428369)

plt.scatter(x, ydata)
plt.plot(x, yfit, color="red")

plt.show()
```
